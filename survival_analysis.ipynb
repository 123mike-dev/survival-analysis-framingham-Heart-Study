# Survival Analysis - Framingham Study

```
import pandas as pd
from lifelines import KaplanMeierFitter, CoxPHFitter
import matplotlib.pyplot as plt

df = pd.read_csv('../data/framingham.csv')
T = df['time']
E = df['event']

kmf = KaplanMeierFitter()
kmf.fit(T, event_observed=E)
kmf.plot()
plt.show()
```
